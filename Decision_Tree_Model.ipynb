```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report,accuracy_score,confusion_matrix
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
url = "https://web.stanford.edu/class/archive/cs/cs109/cs109.1166/stuff/titanic.csv"
titanic_df = pd.read_csv(url)
titanic_df.head()
```

data preprocessing

```python
titanic_df.fillna(titanic_df['Age'].mean(), inplace=True)
titanic_df['Sex'] = titanic_df['Sex'].map({'male': 0, 'female': 1})
```

```python
titanic_df.head(5)
```

```python
X = titanic_df[['Pclass','Sex','Age','Siblings/Spouses Aboard','Parents/Children Aboard','Fare']]
```

```python
Y = titanic_df['Survived']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
```

# Doubt
 How is the maximum depth determined

```python
dt = DecisionTreeClassifier(max_depth=4)
```

```python
dt.fit(X_train,Y_train)
```

```python
prediction = dt.predict(X_test)
```

```python
accuracy_score(Y_test,prediction)
```

```python
confusion_matrix(Y_test,prediction)
```

```python
classification_report(Y_test,prediction)
```

# Predicting the family size using the parent and child columns

```python
url = "https://web.stanford.edu/class/archive/cs/cs109/cs109.1166/stuff/titanic.csv"
df = pd.read_csv(url)
```

```python
# Create family size categories
df['FamilySize'] = df['Siblings/Spouses Aboard'] + df['Parents/Children Aboard']
df['FamilyCategory'] = pd.cut(df['FamilySize'],
                             bins=[-1, 0, 3, 10],
                             labels=[0, 1, 2])
# converting the male female to 1 and 0
df['Sex'] = df['Sex'].map({'male': 0, 'female': 1})
# Predict family category
X_X = df[['Pclass', 'Sex', 'Age', 'Fare', 'Survived']]
Y_Y = df['FamilyCategory']  # 0 means :'Alone', 1 means:'Small', 2 means:'Large'
```

```python
x_train, x_test, y_train , y_test = train_test_split(X_X,Y_Y,test_size=0.2,random_state=42)
```

```python
dt_model = DecisionTreeClassifier()
```

```python
dt_model.fit(x_train,y_train)
```

```python
dt_pred = dt_model.predict(x_test)
```

```python
accuracy_score(y_test,dt_pred)
```

```python
confusion_matrix(y_test,dt_pred)
```

```python
print(classification_report(y_test,dt_pred))
```
